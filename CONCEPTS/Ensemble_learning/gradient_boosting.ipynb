### **Gradient Boosting**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
np.random.seed(42)
x = np.random.rand(100) - 0.5
y = 3 * x[:]**2 + 0.5*np.random.randn(100)
print(f"{x}\n\n")
print(y.mean())
```

```python
df = pd.DataFrame()
df['x'] = x
df['y'] = y

df['pred1'] = y.mean()
df['res1'] = df['y'] - df['pred1']
print(df)
```

```python
plt.scatter(df['x'], df['y'])

plt.title('X vs Y')
plt.xlabel('x------>')
plt.ylabel('y------>')
```

```python
from sklearn.tree import DecisionTreeRegressor
tree1 = DecisionTreeRegressor(max_leaf_nodes = 8)
```

```python
tree1.fit(df['x'].values.reshape(100, 1), df['res1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```
