### **Fetching data from an API**

```python
import numpy as np
import pandas as pd
import requests
```

```python
request = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=dabe3e9e7f149e502a9e1686b39d74b2')
```

```python

for i in range(1, 3):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=dabe3e9e7f149e502a9e1686b39d74b2&page = {}'.format(i))
    temp_data = pd.DataFrame(response.json()['results'])
    df= pd.concat([temp_data], ignore_index = True)
    
    if response.status_code == 200:
        print("Success")
    else:
        print("Error")
        break
    
```
