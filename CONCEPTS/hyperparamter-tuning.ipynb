### **HYPERPARAMETER TUNING:** 
1) GridSearchCV
2) RandomizedSearchCV 


```python
# IMPORTING THE DEPENDENCIES 
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# LOADING THE DATA-SETS

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)

print(data_frame)
 
```

```python
# adding target column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
data_frame.head()
```

```python
# CHECKING IF THERE IS A NULL VALUE OR NOT
data_frame.isnull().sum()
```

```python
x = data_frame.drop(columns= 'label', axis = 1)
y = data_frame['label']
print(x)
print(y)
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

 ### **GridSearchCV**

GridSearchCV is used for determining the best parameters for our model 

```python
model = SVC()

# hyperparametes: 

parameters = {
    
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C' : [1, 5, 10, 20]
    
}

```

```python
classifier = GridSearchCV(model, parameters, cv= 5)

# HERE THE BEST PARAMTERS ARE 'LINEAR' AND C=10
```

```python
classifier.fit(x,y)
```

### **RandomizedSearch_CV**

```python
random = RandomizedSearchCV(model, parameters, cv = 5)
random.fit(x , y)
```
