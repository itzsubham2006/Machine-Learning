```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
class SVM_Classifier():
    def __init__(self, no_of_iteration, learning_rate, lambda_parameter):
        # initializing the hyperparameter 
        self.learning_rate = learning_rate
        self.no_of_iteration = no_of_iteration
        self.lambda_parameter = lambda_parameter
        
    
    def fit(self, x, y):
        # fitting the data-sets into SVM Classifier
        
        self.m, self.n = x.shape
        self.w = np.zeros(self.n)
        self.b = 0
        self.x = x
        self.y = y
        # Implementing the gradient descent:
        for i in range(self.no_of_iteration):
            self.update_weight()
        
    def update_weight(self):
        
        # label encoding: 
        y_label = np.where(self.y<=0, -1, 1)
        
        for index, x_i in enumerate(self.x):
            condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >= 1
            
            if condition== True:
                dw = 2 * self.lambda_parameter * self.w
                db = 0
            else:
                dw = 2 * self.lambda_parameter * self.w - np.dot(y_label[index], x_i)
                
            # updating the bias 
            
            self.w = self.w - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db
            
        
        

    def predict(self, x):
        output =  np.dot(self.w, x) - self.b
        
        predicted_labels = np.sign(output)
        
        y_hat = np.where(predicted_labels<= -1, 0, 1)
        return y_hat
```
